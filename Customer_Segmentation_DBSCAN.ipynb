```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
```

```python
data = {'Customer_ID':[1,2,3,4,5,6,7,8,9,69],
       'Annual_income':[23,54,65,75,23,67,24,75,56,69],
       'Spending_Score':[34,75,87,23,76,88,23,56,32,69]}
```

```python
df = pd.DataFrame(data)
```

```python
x = df[['Annual_income','Spending_Score']]
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
dbscan = DBSCAN(eps=0.8,min_samples = 2)
df['Cluster'] = dbscan.fit_predict(x_scaled)
```

```python
plt.figure(figsize = (6,4))
sns.scatterplot(data = df, x = 'Annual_income', y = 'Spending_Score', hue = 'Cluster', palette = 'Set1',s = 100)
plt.title('Customer Segments with DBSCAN')
plt.legend()
plt.grid(1)
plt.show()
```
